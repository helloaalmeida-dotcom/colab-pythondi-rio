```python
import sqlite3 #importando a biblioteca

#abrindo (se já existir, senão, ele cria) o "caderno de endereço" vulgo DATA BASE/BANCO DE DADOS.
conexao = sqlite3.connect("banco_de_dados.db")
cursor = conexao.cursor() # uma por de entrada para nosso caderno, vamos usa-lo para editar as infos dentro do nosso caderno


print("Conectado ao banco de dados com sucesso")

# anotando as informações que serão lançadas na planilha
data_da_publicacao = "30-10-2025"
fornecedor_produtor = "DUAS RODAS INDUSTRIAL S.A."
cnpj = "84.430.149/0001-09"
marca = "SELECTA"
lote_s = "Todos"
n1 = "Alimentos"
n2 = "Mercearia Seca Doce"
n3 = "Doces"
descricao_do_produto = "Chocolates, recheios e coberturas de diversas linhas (sabores: chocolate, coco, pistache, paçoca, limão, doce de leite, etc.)"
motivo_de_recall = "Rotulagens sem advertências de precaução de alergênicos."
acoes_determinadas = "Recolhimento, Voluntário, Comercialização, Distribuição, Fabricação, Propaganda, Uso"
resolucao = "RESOLUÇÃO-RE Nº 4.262, DE 28 DE OUTUBRO DE 2025"
documento = "RESOLUÇÃO-RE Nº 4.262, DE 28 DE OUTUBRO DE 2025.pdf"
hiperlink = "https://www.in.gov.br/en/web/dou/-/resolucao-re-n-4.262-de-28-de-outubro-de-2025-665141141"

# juntando todas informaç~çoes que anotamos em um aúnica "lsita"
linha = [data_da_publicacao, fornecedor_produtor, cnpj, marca,
        lote_s, n1, n2, n3, descricao_do_produto, motivo_de_recall, acoes_determinadas,
        resolucao, documento, hiperlink] # e aqui também serapara na planilha as informaçoes em colunas

# lançando infos para a planilha
planilha.append_row(linha)
print("Dados enviados para o Google Sheets!")

# criando a estrutura do "caderno", ou seja, montando as colunas do nosso banco de dados
cursor.execute('''
    CREATE TABLE IF NOT EXISTS recall (
        data_da_publicacao TEXT,
        fornecedor_produtor TEXT,
        cnpj TEXT,
        marca TEXT,
        lote_s TEXT,
        n1 TEXT,
        n2 TEXT,
        n3 TEXT,
        descricao_do_produto TEXT,
        motivo_de_recall TEXT,
        acoes_determinadas TEXT,
        resolucao TEXT,
        documento TEXT,
        hiperlink TEXT
    )
''')

# commit salva as informações/alterações no banco de dados
conexao.commit()
print("Dados inseridos no banco de dados com sucesso!")

# encerra a conexao com o banco de dados, segurançazinha para evitar problemas
conexao.close()
print("Dados inseridos no banco de dados com sucesso!")
```
